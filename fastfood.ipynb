```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('fastfood.csv')
df.head(100)
```

```python
plt.figure(figsize=(8,8))
plt.hist(df['restaurant'], bins= 20, color='blue', edgecolor='black')
plt.tight_layout()
plt.xticks()
plt.show()
```

```python
sns.histplot(df['calories'], bins=5, kde=True, color='orange', edgecolor='black')
plt.title("Distribution of Calories Across Fast Food Items")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()
```

```python

df = pd.read_csv("fastfood.csv")

sns.pairplot(df[['calories', 'total_fat', 'protein', 'sodium']], diag_kind='hist')
plt.suptitle("Pairplot: Comparing Nutrients in McDonald's Menu", y=1.02)
plt.show()
```

```python
cal_by_type = df.groupby('salad')['calories'].mean()
plt.pie(cal_by_type, labels=cal_by_type.index, autopct='%1.1f%%', startangle=90)
plt.title("Average Calories by Salad Type")
plt.show()
```

```python
top10_protein = df.sort_values('protein', ascending=False).head(10)
sns.barplot(x='item', y='protein', data=top10_protein, palette='coolwarm')
plt.title("Top 10 Items by Protein Content")
plt.xlabel("Item")
plt.ylabel("Protein (g)")
plt.xticks(rotation=90)
plt.show()
```

```python
df= df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (First 100 McDonald's Items)")
plt.show()
```
